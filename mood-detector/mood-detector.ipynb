```python
def detect_mood(self, user_input: str) -> str:
    """dectect the mood of the user based on input keyword"""
    user_input_lower = user_input.lower()
    mood_score = {mood:0 for mood in self.mood_keywords }

    
    for mood, keywords in self.mood_keyword.item():
        for keyword in keywords:
            if keyword in user_input_lower:
                mood_score[mood] +=1
    detect_mood = max(mood_score, key = mood_score.get)


    if mood_score(detect_mood)==0:
        return 'neutral'
    return detect_mood
```
