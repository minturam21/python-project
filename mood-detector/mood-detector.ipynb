```python
# import the 're' module for regual expression
# 'typing' is a Python module that provides type hints - a way to indicate what type of data a variable, parameter, or return value should be.

from typing import Dict
import random

class MoodDetector:
    def __init__(self):
        self.mood_keywords = {
            'happy': [
                'happy', 'joy', 'excited', 'great', 'awesome', 'amazing', 'wonderful',
                'fantastic', 'glad', 'cheerful', 'delighted', 'thrilled', 'good'
            ],
            'sad': [
                'sad', 'unhappy', 'depressed', 'down', 'miserable', 'heartbroken',
                'disappointed', 'lonely', 'blue', 'upset', 'hurt'
            ],
            'angry': [
                'angry', 'mad', 'furious', 'annoyed', 'irritated', 'frustrated',
                'rage', 'pissed', 'outraged', 'bitter'
            ],
            'anxious': [
                'anxious', 'worried', 'nervous', 'stressed', 'tense', 'uneasy',
                'concerned', 'afraid', 'scared', 'fearful', 'panic'
            ],
            'tired': [
                'tired', 'exhausted', 'sleepy', 'fatigue', 'drained', 'weary',
                'worn out', 'lazy', 'sluggish'
            ],
            'motivated': [
                'motivated', 'inspired', 'determined', 'ambitious', 'driven',
                'energized', 'pumped', 'ready', 'focused'
            ],
            'confused': [
                'confused', 'puzzled', 'uncertain', 'lost', 'bewildered',
                'perplexed', 'unsure', 'doubtful'
            ],
            'grateful': [
                'grateful', 'thankful', 'blessed', 'appreciative', 'fortunate',
                'lucky', 'content'
            ]
        }

        self.mood_quote = {
            'happy': [
                "Keep shining! Your happiness is contagious! âœ¨",
                "Happiness looks good on you! Keep spreading those positive vibes!",
                "What a wonderful mood! May your day continue to be amazing!",
                "Your joy is a gift to the world. Keep smiling! ğŸ˜Š"
            ],
            'sad': [
                "It's okay to feel sad. Remember, storms don't last forever. ğŸŒˆ",
                "Be gentle with yourself. You're doing the best you can.",
                "Even the darkest night will end and the sun will rise. - Victor Hugo",
                "Your feelings are valid. Take all the time you need to heal."
            ],
            'angry': [
                "Take a deep breath. You have the power to choose your response.",
                "Anger is valid, but you are stronger than this feeling.",
                "Let it out, then let it go. Peace is waiting for you. â˜®ï¸",
                "Channel this energy into something positive. You've got this!"
            ],
            'anxious': [
                "Breathe in calm, breathe out worry. You are safe right now.",
                "One step at a time. You don't have to solve everything today.",
                "Your anxiety is lying to you. You are capable and strong. ğŸ’ª",
                "This too shall pass. You've overcome challenges before."
            ],
            'tired': [
                "Rest is not weakness; it's wisdom. Take care of yourself. ğŸŒ™",
                "Your body is asking for what it needs. Listen to it.",
                "Even superheroes need to recharge. It's okay to rest.",
                "Tomorrow is a new day. For now, be kind to yourself."
            ],
            'motivated': [
                "Yes! Ride that wave of motivation! You're unstoppable! ğŸš€",
                "Your determination is inspiring. Keep that fire burning!",
                "Dreams don't work unless you do. And you're DOING IT!",
                "The world needs your energy. Go chase those goals! ğŸ’«"
            ],
            'confused': [
                "It's okay not to have all the answers right now.",
                "Confusion is the first step toward clarity. Keep exploring.",
                "Take your time. Understanding will come. ğŸ§­",
                "Ask questions, seek answers. You'll find your way."
            ],
            'grateful': [
                "Gratitude turns what we have into enough. Beautiful mindset! ğŸ™",
                "Your appreciation for life is inspiring. Keep that perspective!",
                "Gratitude is the healthiest emotion. You're on the right path!",
                "When we focus on blessings, they multiply. Thank you for sharing!"
            ]
        }

    def detect_mood(self, user_input: str) -> str:
        """Detect the mood of the user based on input keywords"""
        user_input_lower = user_input.lower()
        mood_score = {mood: 0 for mood in self.mood_keywords}

        for mood, keywords in self.mood_keywords.items():
            for keyword in keywords:
                if keyword in user_input_lower:
                    mood_score[mood] += 1

        detected_mood = max(mood_score, key=mood_score.get)
        if mood_score[detected_mood] == 0:
            return 'neutral'
        return detected_mood

    def get_quote(self, mood: str) -> str:
        """Get a random quote for the detected mood"""
        if mood == "neutral":
            return "Thank you for sharing! Whatever you are feeling, remember you are not alone. ğŸŒ»"
        return random.choice(self.mood_quote.get(mood, ["Stay strong."]))

    def analyze(self, user_input: str) -> Dict[str, str]:
        """Analyze user input and return mood with quote"""
        mood = self.detect_mood(user_input)
        quote = self.get_quote(mood)
        return {"mood": mood, "quote": quote}
```

```python
def main():
    detector = MoodDetector()
    print("="*60)
    print("ğŸ­ MOOD DETECTOR - Share how you're feeling! ğŸ­")
    print("="*60)
    print("Type 'quit' or 'exit' to end the program\n")

    while True:
        user_input = input("How are you feeling today? ")
        if user_input.lower() in ['quit', 'exit']:
            print("\nTake care! Every feeling is temporary â¤ï¸")
            break
        if not user_input.strip():
            print("Please share something about how you are feeling.")
            continue

        result = detector.analyze(user_input)
        print(f"\nğŸ¯ Detected mood: {result['mood'].upper()}")
        print(f"ğŸ’¬ Message for you: {result['quote']}\n")
        print("-" * 60 + "\n")

main()
```
